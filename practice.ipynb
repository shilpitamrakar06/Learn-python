```python
# datatype
#1 string
x = 'python'
#pyhton function for string 
print(x)
print(type(x))
print(id(x))
print(len(x))
print(max(x))
print(min(x))

#method
x = 'PYthon'
print(x.lower())
print(x.upper())
print(x.title())
print(x.capitalize())
print(x.index('P'))
s = 'this is python'
print(s.find('is'))
print(s.replace('is','IS'))
print(s.startswith('t'))
print(s.endswith('o'))
print(s.split())
      
```

```python
#list 
#python function for list 
l = [ 1,2,3,4,'python']
print(l)
print(type(l))
print(id(l))
print(len(l))
l1 = ['pyhton','java','php','c++','js']
print(max(l1))
print(min(l1))
l2 = [2,4,3,5,6,12,5,9]
print(sum(l2))
```

```python
# method 
l = [2,4,1,5,3,'pyhton','java','php','c++',]
l.append('js')
print(l)
l1 = l.copy()
print(l1)
l1.clear()
print(l1)
print(id(l1))
del l1 
l.extend('java')
print(l)
l.extend('js node')
print(l) 
l.insert(3,'shilpi')
print(l)
l.pop()
print(l)
l.pop(5)
print(l)
l.remove('c++')
print(l)

l.append('rashi')
print(l)
print(l.count('j'))
print(l.count('a'))

l = [2,4,6,9,2,5,7,9,4,32,2]
l.sort()
print(l)
l.reverse()
print(l)
l = [2,4,7,9,4,1,4,5,56,67,3,2,0]
l.sort()
print(l)
l.reverse()
print(l)

         
```
