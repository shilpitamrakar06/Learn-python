```python
n= int(input("Enter any number"))
i,sum=1,0
while(i<=n):
    sum=sum+i
    if i<=(n-1):
        print(i,end= "+")
    else:
        print(i,end= "=")
    i=i+1
print(sum)


    
```

```python
n = int(input("Enter any number: "))
i,sum = n,0
while i >= 1:
    sum = sum + i
    if i>1:
        print(i, end="-")
    else:
        print(i, end="=")
    i = i - 1

print(sum)


    
```
