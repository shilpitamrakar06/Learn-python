```python
##sum==

n = int(input("enter any number"))
sum = 0 
for i in range(1,n+1):
    sum = sum+i
    if i<n:
        print(i,end='+')
    else:
        print(i, end='=')
print(sum)
```

```python
 ## for sub===
n = int(input("enter any number"))
sum = 0 
for i in range(1,n+1):
    sum = sum-i
    if i<n:
        print(i,end='-')
else:
        print(i,end='=')
print(sum)
```
