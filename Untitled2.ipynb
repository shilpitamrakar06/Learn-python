```python
l1 = [1,2,3]
l2 = [4,5,6] 
l3 = [5,6,7]
def add( n1, n2, n3):
    return n1+n2+n3
result = map(add,l1,l2,l3)
#print(result)
print(list(result))
```

```python
l1 = [1,2,3]
l2 = [4,5,6] 
l3 = [5,6,7]
def add( n1, n2, n3):
    return n1+n2 **2+n3**3
result = map(add,l1,l2,l3)
#print(result)
print(list(result))
```

```python
l1 = [1,2,3,4,5,6]
def square(n): 
    return n**2
result = map(square,l1)
print(list(result))
```
