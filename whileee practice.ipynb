```python
n = int(input("Enter any number: "))

original = n
total_digit = 0

# Step 1: Count digits
temp = n
while temp > 0:
    total_digit += 1
    temp = temp // 10

# Step 2: Armstrong calculation
temp = n
sum = 0
while temp > 0:
    last_digit = temp % 10
    sum += last_digit ** total_digit
    temp = temp // 10

# Check result
if sum == original:
    print("Armstrong Number")
else:
    print("Not an Armstrong Number")
```
