```python
## for even number ...

def add(n):
    sum = 0 
    for i in range(2,n+1,2):
        sum = sum +i
        if i<n:
            print(i, end= '+')
        else:
            print(i,end='=')
    print(sum)
p = int(input("enter any number")) 
result = add(p)        
    
    
```

```python
## max function: 
p = "python"
max_ch_ascii =0
for i in p: 
    x = ord(i)
    if max_ch_ascii<x:
        max_ch_ascii = x
print(max_ch_ascii)
print(f'max char is {chr(max_ch_ascii)}')
```

```python
def max_str(p):
    max_ch_ascii =0
    for i in p: 
        x = ord(i)
        if max_ch_ascii<x:
            max_ch_ascii = x
   #print(max_ch_ascii)
    print(f'max char is {chr(max_ch_ascii)}')

max_str(input("Enter any string"))
print("Hello...")
max_str(input("Enter any string"))
```

```python
## for min function :
def min_str(p):
    min_ch_ascii =255
    for i in p: 
        x = ord(i)
        if min_ch_ascii>x:
            min_ch_ascii = x
   #print(max_ch_ascii)
    print(f'min char is {chr(min_ch_ascii)}')

min_str(input("Enter any string"))
print("Hello...")
min_str(input("Enter any string"))
```

```python
# types of arguments:
#1. positional argumnet: 

def new(x,y,z): 
    print('x:',x)
    print('y:',y)
    print('z:',z)
#new()
#new(10)
#new(10,20)
new(10,20,30)
```

```python
#default positional argument:

def new(x=0 ,y=0,z=0): 
    print('x:',x)
    print('y:',y)
    print('z:',z)
#new()
new(10)
new(10,20)
new(10,20,30)
```

```python
#variable length positional argument:

def add(*args):
    print(args)
    print(type(args))
add()
add(1,2,3,4,5)
```

```python

def add (*n):
    sum = 0
    for i in n:
        sum = sum+i
    print(sum)
n = int(input("enter how many number you want to aad:"))
l = []
for i in range(1,n+1):
    value = int(input(f'enter {i}value:')) 
    l.append(value)
print(l)
add(*l)
# add(*l) for unpacking *
```

```python
def add (*n):
    sum = 0
    for i in n:
        for j in i:
           sum = sum+j
    print(sum)
    
n = int(input("enter how many number you want to aad:"))
l = []
for i in range(1,n+1):
    value = int(input(f'enter {i}value:')) 
    l.append(value)
print(l)
add(l)
```
