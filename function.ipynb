```python
## for even number ...

def add(n):
    sum = 0 
    for i in range(2,n+1,2):
        sum = sum +i
        if i<n:
            print(i, end= '+')
        else:
            print(i,end='=')
    print(sum)
p = int(input("enter any number")) 
result = add(p)        
    
    
```

```python
## for oddd number ---
def add(n):
     sum = 0 
     for i in range(1,n+1,2):
         sum = sum+i
         if i<(n-1):
             print(i,end='+')
         else:
             print(i,end= '=')
     print(sum)
p = int(input("enter any value")) 
result = add(p)
```

```python
def odd_no (n):
    i = 1
    while i<=n:
        print(i)
        i = i+2
n = int(input("enter any number"))
odd_no (n)
```

```python
def even_no (n): 
    i = 1
    while i<=n: 
        print(2*i)
        i =i+1
n = int(input("enter any number"))
even_no (n)
    
```

```python
## max function: 
p = "python"
max_ch_ascii =0
for i in p: 
    x = ord(i)
    if max_ch_ascii<x:
        max_ch_ascii = x
print(max_ch_ascii)
print(f'max char is {chr(max_ch_ascii)}')
```

```python
def max_str(p):
    max_ch_ascii =0
    for i in p: 
        x = ord(i)
        if max_ch_ascii<x:
            max_ch_ascii = x
   #print(max_ch_ascii)
    print(f'max char is {chr(max_ch_ascii)}')

max_str(input("Enter any string"))
print("Hello...")
max_str(input("Enter any string"))
```

```python
## for min function :
def min_str(p):
    min_ch_ascii =255
    for i in p: 
        x = ord(i)
        if min_ch_ascii>x:
            min_ch_ascii = x
   #print(max_ch_ascii)
    print(f'min char is {chr(min_ch_ascii)}')

min_str(input("Enter any string"))
print("Hello...")
min_str(input("Enter any string"))
```
