```python
## for even number ...

def add(n):
    sum = 0 
    for i in range(2,n+1,2):
        sum = sum +i
        if i<n:
            print(i, end= '+')
        else:
            print(i,end='=')
    print(sum)
p = int(input("enter any number")) 
result = add(p)        
    
    
```

```python
## for oddd number ---
def add(n):
     sum = 0 
     for i in range(1,n+1,2):
         sum = sum+i
         if i<(n-1):
             print(i,end='+')
         else:
             print(i,end= '=')
     print(sum)
p = int(input("enter any value")) 
result = add(p)
```
